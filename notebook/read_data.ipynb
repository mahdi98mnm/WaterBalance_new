```python
import os
import sys
import datetime
import pickle
import xarray as xr

from google.colab import drive
drive.mount('/content/drive')
```

```python
# Mashhad Limit
LAT_MIN = 35.80
LAT_MAX = 37.11
LON_MIN = 58.31
LON_MAX = 60.14
```

# Create List Files

```python
step = "Daily"
time = "Past"
para = [
    "P",
    # "LWd",
    # "Pres",
    # "RelHum",
    # "SpecHum",
    # "SWd",
    "Temp",
    "Tmax",
    "Tmin",
    # "Wind",
]

result = {}

for p in para:
    filesFolder = os.path.abspath(f"/content/drive/MyDrive/w/{time}/{p}/{step}")
    filesList = os.listdir(filesFolder)
    filesPath = [f"{filesFolder}/{i}" for i in filesList]
    result[f"{p}"] = filesPath

with open(f'/content/drive/MyDrive/w/FilesList.pkl', 'wb') as f:
    pickle.dump(result, f)

```

```python
with open(f'/content/drive/MyDrive/w/FilesList.pkl', 'rb') as f:
    filesList = pickle.load(f)
```

```python
def mask_nc_file(
    nc
):
    with xr.open_dataset(nc) as xr_nc:
        result = xr_nc.sel(
            lat = slice(LAT_MAX, LAT_MIN),
            lon = slice(LON_MIN, LON_MAX),
        )
    return result

result = map(mask_nc_file, filesList["P"])
```
